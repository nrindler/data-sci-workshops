```python
# Prep - Load required packages
%matplotlib inline
import sklearn
from sklearn import datasets
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib
from matplotlib import pyplot as plt
matplotlib.style.use('ggplot')
import numpy as np
import pandas as pd
import random
```

```python
iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
iris.head()
```

### Question 6 - Plot Clusters

Using Python, create a Dendrogram plot that matches the following with the sample data:

- Set chart title: "Hierarchical Clustering of Iris Flowers"
- Set x-axis title: "Observation Number"
- Set y-axis title: "Distance"

```python
random.seed(123)
Z = linkage(iris, method='weighted')
plt.figure(figsize=(25, 10))
# YOUR INPUT HERE  (HINT: SEARCH 'MATPLOTLIB TITLE' ON GOOGLE)
dendrogram(
    Z,
    leaf_font_size=12.,  # font size for the x axis labels
    color_threshold=0.5*max(Z[:,2]) 
)
plt.show()
```

### Question 7 - How many clusters are there, based on the chart above?

Input the number of clusters below to predict Iris flower species using K-Means clustering:

http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

```python
nclust = 3 #YOUR INPUT HERE
```

```python
# Predict flower species using K-Means clustering
random.seed(7)
kmeans = KMeans(n_clusters=nclust, random_state=0).fit(iris)
iris['Species'] = kmeans.predict(iris)
```

```python
# Summarize attributes by Species
iris.groupby('Species').mean()
```
